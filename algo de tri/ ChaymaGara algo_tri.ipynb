```python
temps_d = datetime.now()
temps_triRapide = []
tableau=[]



def TriRapide(v,fin,debut=0):

    def triRapide(v,fin,debut=0):

        def division(v, debut, fin):
            i = (debut-1)		 
            pivot = v[fin]	 

            for j in range(debut, fin):

                if v[j] <= pivot:

                    i = i+1
                    v[i], v[j] = v[j], v[i]

            v[i+1], v[fin] = v[fin], v[i+1]
            return (i+1)

        if len(v) == 1:
            return v
        if debut < fin:
            pivot = division(v, debut, fin)
            triRapide(v, debut, pivot-1)
            triRapide(v, pivot+1, fin)
    temps_f = datetime.now() # prendre le temps apres execution de l'algo
    temps = temps_f-temps_d # difference entre temp de début et temp de fin
    sec = temps.total_seconds()# conversion du temp de difference en seconde
    temps_triRapide.append(float(sec))
    
    
    
borne_inf=20000
borne_sup=100000
pas=5000
for i in range(borne_inf, borne_sup+1, pas):
    tableau.append(i)
    a=rd.randint(100, size=(i))
    pas=len(a)-1
    TriRapide(a,pas)
    print("le tableau trié",a)
```
